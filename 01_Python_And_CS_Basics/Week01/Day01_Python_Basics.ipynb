## 📘 이론 및 개념

###1. 변수란 무엇인가?

정의: 메모리에 저장된 값을 참조하는 이름.
배경: 프로그래밍 언어에서는 값을 임시로 저장하고 사용하기 위해 변수 사용.
예시:

```python
x = 10
name = "Alice"
is_active = True
```

왜 필요한가:  
	•	데이터를 기억하고,  
	•	복잡한 연산이나 로직을 단순화하며,  
	•	코드 재사용성을 높이기 위해.  

| 특징 | 설명 |
|------|------|
| 선언과 동시에 할당 | 별도의 선언 키워드 없이 바로 사용 |
| 자료형 추론 | Python은 자료형을 자동으로 추론 |

### 2. 기본 자료형 종류
| 자료형 | 설명 | 예시 |
|--------|------|------|
| int | 정수 | 1, -5, 0 |
| float | 실수 | 3.14, -0.5 |
| str | 문자열 | "hello", 'world' |
| bool | 불리언 | True, False |

### 3. 출력 함수 print()

```python
print("Hello World")
print(3 + 5)
```
