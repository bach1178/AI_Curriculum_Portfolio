### Day 01 퀴즈

1. Python에서 변수를 선언할 때 필요한 키워드는?  
  a) var  
  b) int  
  c) str  
  d) 없음

2. 다음 중 문자열 자료형에 해당하는 것은?  
  a) 123  
  b) True  
  c) "Hello"  
  d) 3.14

3. 다음 중 print() 함수를 올바르게 사용한 것은?  
  a) print "Hello"  
  b) print("Hello")  
  c) echo("Hello")  
  d) output("Hello")

### 정답 및 해설 (문제 1)
정답: d)  
해설: Python은 별도의 선언 키워드 없이 바로 사용 가능하고 자료형을 자동으로 추론함  

### 정답 및 해설 (문제 2)
정답: c)  
해설: 문자열은 문자를 따옴표(" ")로 감싸서 사용함  

### 정답 및 해설 (문제 3)
정답: b)  
해설: Python에선 출력을 하기 위해선 print()를 사용하여 출력할 변수 및 값을 괄호로 감싸서 출력한다.
